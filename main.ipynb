```python
import torch

from data.dataloaders import get_loaders
from utils.load_config import load_config
from models import build_model          # üëà important change
from trainer import RegressionTrainer


def main():
    # ---------------------------
    # LOAD CONFIG
    # ---------------------------
    config_name = "multi_model" 
    config = load_config(f"{config_name}.yml")

    config["device"] = "cuda" if torch.cuda.is_available() else "cpu"

    # ---------------------------
    # LOG CONFIG
    # ---------------------------
    print("=" * 60)
    print("Training Configuration:")
    print(f"  Experiment Name : {config['name']}")
    print(f"  Dataset         : {config['Dataset']}")
    print(f"  Modality        : {config['modality']}")
    print(f"  Batch Size      : {config['batch_size']}")
    print(f"  Epochs          : {config['epochs']}")
    print(f"  Subset Percent  : {config['subset_percent']}")
    print(f"  Device          : {config['device']}")
    print("=" * 60)

    # ---------------------------
    # LOAD DATA
    # ---------------------------
    print("\nLoading data...")
    train_loader, val_loader, test_loader = get_loaders(
        name=config["Dataset"],
        batch_size=config["batch_size"],
        subset_pct=config["subset_percent"],
        modality=config["modality"],
    )

    # ---------------------------
    # BUILD MODEL
    # ---------------------------
    print("\nInitializing model...")
    model = build_model(config)
    model = model.to(config["device"])

    # ---------------------------
    # TRAINER
    # ---------------------------
    save_dir = f"./checkpoints/{config['name']}"
    if config["modality"] != "multi":
        save_dir = f"{save_dir}/{config['modality']}"

    print("\nInitializing trainer...")
    trainer = RegressionTrainer(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        test_loader=test_loader,
        device=config["device"],
        save_dir=save_dir,
        config=config,
    )

    # ---------------------------
    # TRAIN
    # ---------------------------
    print("\n" + "=" * 60)
    print("Starting training...")
    print("=" * 60 + "\n")

    trainer.train(epochs=config["epochs"])

    print("\n" + "=" * 60)
    print("Training completed!")
    print("=" * 60)

if __name__ == "__main__":
    main()
```

```python
import torch
import matplotlib.pyplot as plt
from pathlib import Path

# -------------------------
# CONFIG: MODEL CHECKPOINTS
# -------------------------
MODEL_CHECKPOINTS = {
    "RNA": "checkpoints/uni_rna/RNA/best_model.pt",
    "DNA": "checkpoints/uni_dna/DNA/best_model.pt",
    "Protein": "checkpoints/uni_protein/Protein/best_model.pt",
    "Multi": "checkpoints/multi_model/best_model.pt",
}

# -------------------------
# LOAD SPEARMAN HISTORIES
# -------------------------
spearman_histories = {}

for model_name, ckpt_path in MODEL_CHECKPOINTS.items():
    ckpt_path = Path(ckpt_path)

    if not ckpt_path.exists():
        print(f"‚ùå Missing checkpoint for {model_name}: {ckpt_path}")
        continue

    checkpoint = torch.load(ckpt_path, map_location="cpu")
    history = checkpoint.get("history", {})

    train_s = history.get("train_spearman", [])
    val_s   = history.get("val_spearman", [])

    spearman_histories[model_name] = {
        "train": train_s,
        "val": val_s
    }

    # ---- Print final values ----
    if len(val_s) > 0:
        print(
            f"{model_name:8s} | "
            f"Final Train œÅ = {train_s[-1]:.4f} | "
            f"Final Val œÅ = {val_s[-1]:.4f}"
        )

# -------------------------
# PLOT SPEARMAN CURVES
# -------------------------
plt.figure(figsize=(10, 6))

for model_name, data in spearman_histories.items():
    epochs = range(1, len(data["val"]) + 1)
    plt.plot(epochs, data["val"], marker="o", label=f"{model_name} (Val)")

plt.xlabel("Epoch")
plt.ylabel("Spearman œÅ")
plt.title("Validation Spearman Correlation Across Models")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("spearman_comparison.png", dpi=300)
plt.show()
```

```python
import torch
import gc
import numpy as np
from scipy.stats import spearmanr
from models import build_model
from data.dataloaders import get_loaders


def evaluate_single_model(modality, checkpoint_path):
    device = "cpu"
    max_len = 1022   # must match training

    ckpt = torch.load(checkpoint_path, map_location=device)
    state = ckpt["model_state_dict"]

    # ---------------------------
    # Infer architecture params
    # ---------------------------
    if modality.lower() == "multi":
        projection_dim = state["proj_dna.proj.0.weight"].shape[0]
        dDNA  = state["proj_dna.proj.0.weight"].shape[1]
        dRNA  = state["proj_rna.proj.0.weight"].shape[1]
        dProt = state["proj_prot.proj.0.weight"].shape[1]

        model = build_model({
            "modality": "multi",
            "device": device,
            "max_len": max_len,
            "projection_dim": projection_dim,
            "dDNA": dDNA,
            "dRNA": dRNA,
            "dProt": dProt,
        }).to(device)

    else:
        # unimodal
        projection_dim = state["prediction_head.net.0.weight"].shape[0]

        model = build_model({
            "modality": modality,
            "device": device,
            "max_len": max_len,
            "projection_dim": projection_dim,
        }).to(device)

    # ---------------------------
    # Load weights
    # ---------------------------
    model.load_state_dict(state)
    model.eval()

    # ---------------------------
    # Load test data
    # ---------------------------
    _, _, test_loader = get_loaders(
        name="fungal_expression",
        batch_size=1,
        modality=modality
    )

    preds, targets = [], []

    with torch.no_grad():
        for batch in test_loader:
            if isinstance(batch, dict):
                y = batch["label"].item()
                out = model(batch).item()
            else:
                x, y = batch
                out = model(x).item()
                y = y.item()

            preds.append(out)
            targets.append(y)

    rho, pval = spearmanr(targets, preds)
    return rho, pval


rho, pval = evaluate_single_model(
    modality="DNA",
    checkpoint_path="checkpoints/uni_dna/DNA/best_model.pt"
)
print(f"DNA Spearman: {rho:.4f}, p={pval:.2e}")


rho, pval = evaluate_single_model(
    modality="Protein",
    checkpoint_path="checkpoints/uni_protein/Protein/best_model.pt"
)
print(f"Protein Spearman: {rho:.4f}, p={pval:.2e}")

rho, pval = evaluate_single_model(
    modality="RNA",
    checkpoint_path="checkpoints/uni_rna/RNA/best_model.pt"
)
print(f"RNA Spearman: {rho:.4f}, p={pval:.2e}")


rho, pval = evaluate_single_model(
    modality="multi",
    checkpoint_path="checkpoints/multi_model/best_model.pt"
)
print(f"Multi Spearman: {rho:.4f}, p={pval:.2e}")
```

```python
import torch
import numpy as np
from scipy.stats import spearmanr
from models import build_model
from data.dataloaders import get_loaders


def evaluate_single_model(modality, checkpoint_path):
    device = "cpu"

    # ‚ö†Ô∏è MUST MATCH TRAINING CONFIG
    max_len = 1022            # or 1022 if that‚Äôs what you trained with
    projection_dim = 256     # or whatever you used in training

    _, _, test_loader = get_loaders(
        name="fungal_expression",
        batch_size=1,
        modality=modality
    )

    model = build_model({
        "modality": modality,
        "device": device,
        "max_len": max_len,
        "projection_dim": projection_dim
    }).to(device)

    ckpt = torch.load(checkpoint_path, map_location=device)
    model.load_state_dict(ckpt["model_state_dict"])
    model.eval()

    preds, targets = [], []

    with torch.no_grad():
        for batch in test_loader:
            if isinstance(batch, dict):
                y = batch["label"].item()
                out = model(batch).item()
            else:
                x, y = batch
                out = model(x).item()
                y = y.item()

            preds.append(out)
            targets.append(y)

    rho, pval = spearmanr(targets, preds)

    # cleanup (important for Jupyter)
    del model, ckpt
    import gc
    gc.collect()
    torch.cuda.empty_cache()

    return rho, pval

rho, pval = evaluate_single_model(
    modality="RNA",
    checkpoint_path="checkpoints/uni_rna/RNA/best_model.pt"
)
print(f"RNA Spearman: {rho:.4f}, p={pval:.2e}")
```
